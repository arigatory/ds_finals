–ú—ã –±—É–¥–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –¥–≤—É—Å—Ç–æ—Ä–æ–Ω–Ω–∏–π –∫—Ä–∏—Ç–µ—Ä–∏–π, —Ç–∞–∫ –∫–∞–∫ –º—ã —Ö–æ—Ç–∏–º —É–∑–Ω–∞—Ç—å, –æ—Ç–ª–∏—á–∞—é—Ç—Å—è –ª–∏ —Å—Ä–µ–¥–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è –º–µ–∂–¥—É —Å—Ä–∞–≤–Ω–∏–≤–∞–µ–º—ã–º–∏ –≥—Ä—É–ø–ø–∞–º–∏ –≤ –ª—é–±–æ–º –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏–∏.

–ü–æ—Ä–æ–≥–æ–≤–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ alpha –∑–∞–¥–∞–¥–∏–º —Ä–∞–≤–Ω—ã–º 0.05.

–ù—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ (H0) –¥–ª—è –ø–µ—Ä–≤–æ–π –≥–∏–ø–æ—Ç–µ–∑—ã –±—É–¥–µ—Ç –∑–≤—É—á–∞—Ç—å —Ç–∞–∫: —Å—Ä–µ–¥–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º Xbox One –∏ PC —Ä–∞–≤–Ω—ã.
–ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ (H1): —Å—Ä–µ–¥–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º Xbox One –∏ PC –æ—Ç–ª–∏—á–∞—é—Ç—Å—è.

```python
xbox_one_user_ratings = df[df['platform'] == 'XOne']['user_score'].dropna()
pc_user_ratings = df[df['platform'] == 'PC']['user_score'].dropna()
```

```python
from scipy import stats
```

–î–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ –ø–µ—Ä–≤–æ–π –≥–∏–ø–æ—Ç–µ–∑—ã –º—ã –±—É–¥–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å t-—Ç–µ—Å—Ç –°—Ç—å—é–¥–µ–Ω—Ç–∞ –¥–ª—è –¥–≤—É—Ö –≤—ã–±–æ—Ä–æ–∫ —Å —Ä–∞–≤–Ω—ã–º–∏ –¥–∏—Å–ø–µ—Ä—Å–∏—è–º–∏, —Ç–∞–∫ –∫–∞–∫ –º—ã —Ö–æ—Ç–∏–º —Å—Ä–∞–≤–Ω–∏—Ç—å —Å—Ä–µ–¥–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è –¥–≤—É—Ö –≥–µ–Ω–µ—Ä–∞–ª—å–Ω—ã—Ö —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç–µ–π (—Å—Ä–µ–¥–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º Xbox One –∏ PC).

```python
alpha = 0.05
results = stats.ttest_ind(xbox_one_user_ratings, pc_user_ratings, equal_var=True)
print('p-value:', results.pvalue)

if results.pvalue < alpha:
    print('–û—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É')
else:
    print('–ù–µ —É–¥–∞–ª–æ—Å—å –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É')
```

–î–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ –≤—Ç–æ—Ä–æ–π –≥–∏–ø–æ—Ç–µ–∑—ã (–æ —Ä–∞–∑–ª–∏—á–∏–∏ —Å—Ä–µ–¥–Ω–∏—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏—Ö —Ä–µ–π—Ç–∏–Ω–≥–æ–≤ –¥–ª—è –∂–∞–Ω—Ä–æ–≤ Action –∏ Sports), —Ç–∞–∫–∂–µ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ —Å—Ñ–æ—Ä–º–∏—Ä–æ–≤–∞—Ç—å –¥–≤–µ –≤—ã–±–æ—Ä–∫–∏ –∏–∑ –¥–∞–Ω–Ω—ã—Ö –∏—Å—Ö–æ–¥–Ω–æ–≥–æ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞:

–ù—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ (H0) –¥–ª—è –≤—Ç–æ—Ä–æ–π –≥–∏–ø–æ—Ç–µ–∑—ã –±—É–¥–µ—Ç –∑–≤—É—á–∞—Ç—å —Ç–∞–∫: —Å—Ä–µ–¥–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ –∂–∞–Ω—Ä–æ–≤ Action –∏ Sports —Ä–∞–≤–Ω—ã.
–ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ (H1): —Å—Ä–µ–¥–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ –∂–∞–Ω—Ä–æ–≤ Action –∏ Sports –æ—Ç–ª–∏—á–∞—é—Ç—Å—è.

–î–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ –≤—Ç–æ—Ä–æ–π –≥–∏–ø–æ—Ç–µ–∑—ã –º—ã –±—É–¥–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å t-—Ç–µ—Å—Ç –°—Ç—å—é–¥–µ–Ω—Ç–∞ –¥–ª—è –¥–≤—É—Ö –≤—ã–±–æ—Ä–æ–∫ —Å –Ω–µ —Ä–∞–≤–Ω—ã–º–∏ –¥–∏—Å–ø–µ—Ä—Å–∏—è–º–∏, —Ç–∞–∫ –∫–∞–∫ –º—ã —Ö–æ—Ç–∏–º —Å—Ä–∞–≤–Ω–∏—Ç—å —Å—Ä–µ–¥–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è –¥–≤—É—Ö –≥–µ–Ω–µ—Ä–∞–ª—å–Ω—ã—Ö —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç–µ–π (—Å—Ä–µ–¥–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ –∂–∞–Ω—Ä–æ–≤ Action –∏ Sports), –Ω–æ –¥–∏—Å–ø–µ—Ä—Å–∏–∏ —ç—Ç–∏—Ö –≥–µ–Ω–µ—Ä–∞–ª—å–Ω—ã—Ö —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç–µ–π –º–æ–≥—É—Ç –±—ã—Ç—å –Ω–µ —Ä–∞–≤–Ω—ã.

```python
alpha = 0.05
results = stats.ttest_ind(action_user_ratings, sports_user_ratings, equal_var=False)
print('p-value:', results.pvalue)

if results.pvalue < alpha:
    print('–û—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É')
else:
    print('–ù–µ —É–¥–∞–ª–æ—Å—å –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É')
```
